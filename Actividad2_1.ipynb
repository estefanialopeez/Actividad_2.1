```python
df=pd.read_csv('Ventas_totales.csv')
df.head()
```

```python
nulos=df.isnull().sum()
nulos
#valores nulos en columnas:

```

```python
df2=df.copy()
#Verificar el tipo de datos
df2.info()
```

```python
#sustitucion por medio de la media

df2['salon_ventas']=df2['salon_ventas'].fillna(round(df['salon_ventas'].mean(),1))
df2['tarjetas_debito']=df2['tarjetas_debito'].fillna(round(df['tarjetas_debito'].mean(),1))
df2['tarjetas_credito']=df2['tarjetas_credito'].fillna(round(df['tarjetas_credito'].mean(),1))

df2['otros_medios']=df2['otros_medios'].fillna(round(df['otros_medios'].mean(),1))
df2['ventas_totales_grupo_articulos']=df2['ventas_totales_grupo_articulos'].fillna(round(df['ventas_totales_grupo_articulos'].mean(),1))
df2['subtotal_ventas_alimentos_bebidas']=df2['subtotal_ventas_alimentos_bebidas'].fillna(round(df['subtotal_ventas_alimentos_bebidas'].mean(),1))
nulos=df2.isnull().sum()
nulos
```

```python
#sustitucion por eliminacion
#se uso este metodo en las columnas donde solo se encuentra 1 valor nulo
df3=df2.copy()

columnas_a_procesar = ['bebidas', 'panaderia', 'lacteos', 'carnes', 'verduleria_fruteria', 'alimentos_preparados_rotiseria','electronicos_articulos_hogar']

# Elimina las filas que contienen datos nulos en las columnas especificadas
df3 = df2.dropna(subset=columnas_a_procesar)

df3.isnull().sum()
```

```python
df4=df3.copy()
#sustitucion por medio de la mediana
df4['almacen']=df4['almacen'].fillna(round(df3['almacen'].median(),1))
df4['indumentaria_calzado_textiles_hogar']=df4['indumentaria_calzado_textiles_hogar'].fillna(round(df3['indumentaria_calzado_textiles_hogar'].median(),1))
df4['otros']=df4['otros'].fillna(round(df3['otros'].median(),1))
```

```python
Valores_nulos=df4.isnull().sum()
print('LOS VALORES NULOS POR COLUMNA SON:\n',Valores_nulos)
Total_Valores_nulos=df4.isnull().sum().sum()

print('EL TOTAL DE VALORES NULOS SON: ', Total_Valores_nulos)
```

```python
df4.to_csv('Ventas_sin_nulos.csv')
```
